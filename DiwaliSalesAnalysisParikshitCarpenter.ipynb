```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
%matplotlib notebook
```

```python
data=pd.read_csv("D:\Analytics\Datasets\Diwali Rishabh Mishra\Diwali Sales Data.csv", encoding='latin1')
data.head()
```

```python
df=data.copy()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df['Amount']=df['Amount'].astype(int)
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status':'Shaadi'}, inplace=True)
```

```python
# describe() method describe the data we have, it tells us mean, median, standard deviation etc..etc...
df.describe()
```

```python
df.head()
```

# Exploratory Data Analysis

```python
import seaborn as sns
```

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Summarize the categorical data is by using the value_counts() function.
df.value_counts().to_frame()
```

```python
# plotting a bar chart for gender vs total amount 
sales_gen=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
ax=sns.barplot(x='Gender', y='Amount', data=sales_gen)
for bars in ax.containers:
    ax.bar_label(bars)
```

From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

### Age

```python
ax=sns.countplot(x='Age Group', hue='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax=sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

From above graphs we can see that most of the buyers are of age group between 26-35 yrs female.

### State

```python
sales_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
plt.figure(figsize=(15,5))
sns.barplot(data=sales_state, x='State', y='Orders')
plt.xticks(rotation=90)
plt.tight_layout()
```

```python
# Total Amount/Sales from top 10 states
state_amt=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sales_state['Total Amount']=state_amt['Amount']
```

```python
sales_state=sales_state[:10]
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='State', y='Total Amount', data=sales_state)
```

From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively

### Marital Status

```python
df.rename(columns={'Shaadi':'Marital Status'}, inplace=True)
```

```python
ax=sns.countplot(data=df, x='Marital Status')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_mar=df.groupby(['Marital Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.figure(figsize=(15,5))
ax=sns.barplot(x='Marital Status', y='Amount',hue='Gender', data=sales_mar)
for bars in ax.containers:
    ax.bar_label(bars)
```

From above graphs we can see that most of the buyers are married (women) and they have high purchasing power

### Occupation

```python
df['Occupation'].unique()
```

```python
plt.figure(figsize=(10,8))
ax=sns.countplot(x='Occupation', data=df)
plt.tight_layout()
plt.xticks(rotation=90)
plt.title('Sectors Worker Count')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occup=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plt.figure(figsize=(15,8))
ax=sns.barplot(x='Occupation', y='Amount', data=sales_occup)
for bars in ax.containers:
    ax.bar_label(bars)
plt.tight_layout()
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector.

### Product Category

```python
df['Product_Category'].unique()
```

```python
plt.figure(figsize=(10,8))
plt.tight_layout()
ax=sns.countplot(x='Product_Category', data=df)
plt.xticks(rotation=90)
plt.title('Sold Products Count')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_prod=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
plt.figure(figsize=(15,8))
ax=sns.barplot(x="Product_Category", y="Amount", data=sales_prod.head(5))
plt.tight_layout()
# plt.xticks(rotation=90)

for bars in ax.containers:
    ax.bar_label(bars)
```

From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category.

```python
# %matplotlib notebook
%matplotlib inline
```

```python
sales_prodID=df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
plt.figure(figsize=(20,6))
ax=sns.barplot(x='Product_ID', y='Orders', data=sales_prodID.head(20))
plt.tight_layout()

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Top 10 most sold products
fig1, ax1=plt.subplots(figsize=(15,5))
top_products=df.groupby(['Product_ID'])['Orders'].sum().nlargest(10).sort_values(ascending=False).reset_index()
ax=sns.barplot(x='Product_ID', y='Orders', ax=ax1, data=top_products)
```

The variable fig1 represents the figure object, which is the top-level container for all the plot elements. It allows you to modify the overall properties of the plot, such as the size, title, and saving the plot to a file.

The variable ax1 represents the axes object. It is the actual plot area where you can plot your data and customize the individual axes properties, such as the labels, ticks, and grid.
